```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("new_File1.csv")
df
```

```python
df["Gender"].value_counts().plot(kind="bar", color='skyblue')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

```

```python
df["Age"].plot(kind="hist", bins=10, color='orange', edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Start Date"] = pd.to_datetime(df["Start Date"])
df["End Date"] = pd.to_datetime(df["End Date"])
```

```python
df["duration_days"] = (df["End Date"] - df["Start Date"]).dt.days
df
```

```python
mean = df["duration_days"].mean()
print(f"The average duration of days of studenrs is {average}")
```

```python


median = df["duration_days"].median()
print(f"The average duration of days of studenrs is {average}")
```

```python
max = df["duration_days"].max()
print(f"The average duration of days of studenrs is {average}")
```

```python
min = df["duration_days"].min()
print(f"The average duration of days of studenrs is {average}")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df.index, y=df["duration_days"], label = 'Normal data')
plt.axhline(mean, color = 'green', linestyle = '--', label = 'mean')
plt.axhline(median, color = 'orange', linestyle = '--', label = 'median')
plt.axhline(max, color = 'purple', linestyle = '--', label = 'max')
plt.axhline(min, color = 'black', linestyle = '--', label = 'min')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df["Start Date"])
plt.title("Start Date Trends")
plt.xlabel("Index Value")
plt.ylabel("Start Date")
plt.show()
```

```python
df["School"].value_counts().plot(kind="pie", autopct='%1.1f%%', startangle=90)
plt.title("popular Universities")
plt.ylabel("")
plt.show()
```

```python
df["School"].value_counts().plot(kind="bar", color='skyblue')
plt.title("Popular Universities")
plt.xlabel("School")
plt.ylabel("Count")
plt.show()
```

```python


education_counts = df['Highest Level of Education Completed'].value_counts()

# 2. Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    education_counts,
    labels=education_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff9999'],  # Custom colors
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}  # Add white borders
)

plt.title('Distribution of Highest Education Levels')
plt.axis('equal')  # Ensures pie is drawn as a circle

plt.show()
```

```python
cross_tab = pd.crosstab(df['School d\'pt'], df['Stack'])
plt.figure(figsize=(12,8))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('School Department vs Track at Skye 8')
plt.xlabel('Skye8 Role')
plt.ylabel('School Department')
plt.xticks(rotation=45)
plt.show()
```
